```python
Part 1: Theoretical Queltionk

1. What are the different data structures used in TensorFlow? Give some examples.

ANS- TensorFlow primarily uses two main data structures: tensors and graphs.

1. Tensors: Tensors are the fundamental data structures in TensorFlow. They are multi-dimensional arrays or n-dimensional matrices that 
            represent the data flowing through the computational graph. Tensors can be scalars (0-dimensional), vectors (1-dimensional), 
            matrices (2-dimensional), or higher-dimensional arrays. Examples of tensors include a single number, an image, or a batch of 
            images.

2. Graphs: TensorFlow uses a computational graph to represent the operations and the flow of data between operations. The graph consists 
           of nodes that represent operations and edges that represent the flow of data, which are tensors. The graph defines the structure 
           of the computation and allows for efficient parallel execution on different devices.
        

        
2. How does the TensorFlow constant differ from a TensorFlow variable? Explain with an example.

ANS- In TensorFlow, a constant is a type of tensor whose value cannot be changed once it is defined, while a variable is a type of tensor 
     whose value can be modified during the execution of a TensorFlow graph.
    
Example of a TensorFlow constant:
        
import tensorflow as tf

# Define a constant tensor
a = tf.constant(5)

# Attempt to change the value of the constant
# This will raise an error
a.assign(10)  # Raises an error: 'tensorflow.python.framework.ops.EagerTensor' object has no attribute 'assign'

In the example above, a is a TensorFlow constant with a value of 5. If we attempt to assign a new value to the constant using the 
assign() method, it raises an error because the value of a constant cannot be modified.



Example of a TensorFlow variable:
    
import tensorflow as tf

# Define a variable tensor
b = tf.Variable(5)

# Assign a new value to the variable
b.assign(10)

# Print the updated value of the variable
print(b.numpy())  # Output: 10

In this example, b is a TensorFlow variable with an initial value of 5. We can use the assign() method to modify the value of the 
variable. In this case, we assign a new value of 10 to the variable, and when we print the value, it shows the updated value of 10.




3. Describe the process of matrix addition, multiplication, and element-wise operations in TensorFlow.

ANS- Matrix Addition: In TensorFlow, matrix addition can be performed using the tf.add() function or the + operator. The two matrices 
     being added must have compatible shapes, and the corresponding elements are added together. For example:
        
import tensorflow as tf

# Define two matrices
A = tf.constant([[1, 2], [3, 4]])
B = tf.constant([[5, 6], [7, 8]])

# Perform matrix addition
C = tf.add(A, B)  # or C = A + B

# Print the result
print(C.numpy())  # Output: [[6  8]
                  #          [10 12]]

    
    
Matrix Multiplication: TensorFlow provides the tf.matmul() function or the @ operator for matrix multiplication. The matrices being 
multiplied must have compatible shapes, where the number of columns in the first matrix must be equal to the number of rows in the 
second matrix. For example:
    
import tensorflow as tf

# Define two matrices
A = tf.constant([[1, 2], [3, 4]])
B = tf.constant([[5, 6], [7, 8]])

# Perform matrix multiplication
C = tf.matmul(A, B)  # or C = A @ B

# Print the result
print(C.numpy())  # Output: [[19 22]
                  #          [43 50]]

    
    
Element-wise Operations: TensorFlow supports element-wise operations on tensors, where each element of the input tensors is operated 
on individually. Element-wise operations can be performed using the standard arithmetic operators or the corresponding TensorFlow 
functions. For example, element-wise addition can be done as follows:
    
import tensorflow as tf

# Define two tensors
A = tf.constant([1, 2, 3])
B = tf.constant([4, 5, 6])

# Perform element-wise addition
C = tf.add(A, B)  # or C = A + B

# Print the result
print(C.numpy())  # Output: [5 7 9]

These operations can be performed on tensors of any shape, as long as the shapes are compatible according to the rules of matrix 
addition, multiplication, or element-wise operations.
```
